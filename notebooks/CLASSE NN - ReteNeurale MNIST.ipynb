```
  def train(self, X_train, y_train, n_pixel):
    #X
    X_train = X_train.reshape(X_train.shape[0], n_pixel)
    X_train = X_train / 255
    X_train= X_train.astype('float32')
    #y
    y_train = np_utils.to_categorical(y_train)
    return X_train, y_train
  
  def test(self, X_test, y_test, n_pixel):
    #X
    X_test = X_test.reshape(X_test.shape[0], n_pixel)
    X_test = X_test / 255
    X_test= X_test.astype('float32')
    #y
    y_test = np_utils.to_categorical(y_test)
    return X_test, y_test
  
  def rete(self, layer, X_train, X_test, y_train, y_test):
    
    #creo modello
    model= Sequential()
       
    #primo layer
    neuroni=input('Inserisci il numero di neuroni: ')
    neuroni=int(layer)
    model.add(Dense(neuroni, input_dim=n_pixel, activation='relu'))
    
    i=2
    #altri layer
    while(i<layer):
      neuroni=input('Inserisci il numero di neuroni: ')
      neuroni=int(layer)
      model.add(Dense(neuroni, activation='relu'))
      i=i+1

    #layer di output
    neuroni=input('Inserisci il numero di neuroni: ')
    neuroni=int(layer)
    model.add(Dense(neuroni, activation='softmax'))

    #Compilo modello
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    #alleno
    progress=model.fit(X_train, y_train, validation_data=(X_test, y_test), validation_split=0.3, epochs=10, batch_size=256, shuffle=True)
    
```

```
#IMPORTO DATASET â†’ MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X = np.concatenate((X_train, X_test))
y = np.concatenate((y_train, y_test))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```
classe=NN()
n_pixel=X_train.shape[1]*X_train.shape[2]

#train
(X_train, y_train)=classe.train(X_train, y_train, n_pixel)
#X_train.shape, y_train.shape

#test
(X_test, y_test)=classe.test(X_test, y_test, n_pixel)
#X_test.shape,y_test.shape
```

```
#PERDITA ACCURATEZZA
plt.plot(progress.history['accuracy'])
plt.plot(progress.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
